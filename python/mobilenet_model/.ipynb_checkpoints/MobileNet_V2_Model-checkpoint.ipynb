# 基于mobilenet v2 model通过迁移学习构筑新的模型，并继续图片预测

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import tensorflow as tf
import keras
import pickle
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image_dataset_from_directory
from keras import models
from keras.models import load_model
from sklearn.metrics import accuracy_score
from keras import optimizers
from keras.preprocessing.image import ImageDataGenerator
```

```python
if os.getcwd().split('/')[-1] == "mobilenet_model":
    os.chdir("..")
PATH = os.path.join(os.getcwd(), 'hand_sign_digit_data')

train_dir = os.path.join(PATH, 'train')
validation_dir = os.path.join(PATH, 'validation')

classes = [
    'zero', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight',
    'nine'
]

class_num = len(classes)
BATCH_SIZE = 32
epochs_num = 30
epochs_num_fine_tuning = 15
IMG_SIZE = (224, 224)
```

```python
train_dataset = image_dataset_from_directory(train_dir,
                                             label_mode='categorical',
                                             shuffle=True,
                                             batch_size=BATCH_SIZE,
                                             image_size=IMG_SIZE)
```

```python
validation_dataset = image_dataset_from_directory(validation_dir,
                                                  label_mode='categorical',
                                                  shuffle=True,
                                                  batch_size=BATCH_SIZE,
                                                  image_size=IMG_SIZE)
```

```python
val_batches = tf.data.experimental.cardinality(validation_dataset)
# 验证集中一共有32个批次的batch数据（一个batch里面有32张图片），32个批次分成26和6，2个部分（验证和测试）
test_dataset = validation_dataset.take(val_batches//5)
validation_dataset = validation_dataset.skip(val_batches//5)
```

```python
print('Number of validation batches: %d' % tf.data.experimental.cardinality(validation_dataset))
print('Number of test batches: %d' % tf.data.experimental.cardinality(test_dataset))
```

```python
AUTOTUNE = tf.data.experimental.AUTOTUNE

train_dataset = train_dataset.prefetch(buffer_size=AUTOTUNE)
validation_dataset = validation_dataset.prefetch(buffer_size=AUTOTUNE)
test_dataset = test_dataset.prefetch(buffer_size=AUTOTUNE)
```

### Data Augmentation processing

```python
# 数据扩张需要注意，不要设定的过多，这样会降低准确率，拟合会很费劲，下面的4个设定使用就可以了
# 水平翻转，旋转，图片上下shift10%，左右shift10%，独立的调整每个图片每个通道的对比度
data_augmentation=tf.keras.Sequential([
#     水平 垂直 翻转
    tf.keras.layers.experimental.preprocessing.RandomFlip(mode='horizontal'),
#     旋转
    tf.keras.layers.experimental.preprocessing.RandomRotation(0.1),
#     缩放
#     tf.keras.layers.experimental.preprocessing.RandomZoom(-0.1, 0.1),
    #     图片上下shift10%，左右shift10%
    tf.keras.layers.experimental.preprocessing.RandomTranslation(height_factor=0.1, width_factor=0.1),
#     独立的调整每个图片每个通道的对比度
    tf.keras.layers.experimental.preprocessing.RandomContrast(factor=0.1),
])
```

```python
image, label = next(iter(validation_dataset))
len(image)
```

```python
plt.figure(figsize=(3, 3))
first_image = image[0]
plt.subplot(1,1,1)
plt.imshow(first_image.numpy().astype("uint8"))
plt.title("original")
plt.axis('off')
plt.show()
plt.figure(figsize=(10, 10))
for i in range(9):
    ax = plt.subplot(3, 3, i + 1)
    augmented_image = data_augmentation(tf.expand_dims(first_image, 0))
    plt.imshow(augmented_image[0] / 255)
    plt.title("Data Augmentation")
    plt.axis('off')
```

```python
preprocess_input = tf.keras.applications.mobilenet_v2.preprocess_input
```

```python
# 通过preprocess_input处理，数据变成了-1到1之间的数值
a=preprocess_input(images[0])
print(np.min(a),np.max(a))
# 原始图片是0到255之间的数值
np.min(images[0]),np.max(images[0])
```

```python
rescale = tf.keras.layers.experimental.preprocessing.Rescaling(1./127.5, offset= -1)
```

```python
IMG_SHAPE = IMG_SIZE + (3,)
base_model = tf.keras.applications.MobileNetV2(input_shape=IMG_SHAPE,
                                               include_top=False,
                                               weights='imagenet')
```

```python
image_batch, label_batch = next(iter(train_dataset))
feature_batch = base_model(image_batch)
print(feature_batch.shape)
```

```python
base_model.trainable = False
```

```python
base_model.summary()
```

```python
global_average_layer = tf.keras.layers.GlobalAveragePooling2D()
feature_batch_average = global_average_layer(feature_batch)
print(feature_batch_average.shape)
```

```python
prediction_layer = tf.keras.layers.Dense(10,activation='softmax')
prediction_batch = prediction_layer(feature_batch_average)
print(prediction_batch.shape)
```

```python
prediction_batch[0]
```

```python
# 使用functional API链接各个层
inputs = tf.keras.Input(shape=(IMG_SIZE[0], IMG_SIZE[1], 3))
x = data_augmentation(inputs)
# 进行数据预处理，每种模型都有自己符合的预处理方式，这个是mobilenet v2的预处理
x = preprocess_input(x)
# training=False表示推论模式，training=True表示训练模式
x = base_model(x, training=False)
x = global_average_layer(x)
x = tf.keras.layers.Dropout(0.2)(x)
outputs = prediction_layer(x)
model = tf.keras.Model(inputs, outputs)
```

```python
model.summary()
```

```python
base_learning_rate = 0.0001
model.compile(optimizer=tf.keras.optimizers.Adam(lr=base_learning_rate),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=4)
history = model.fit(train_dataset,
                    epochs=epochs_num,
                    validation_data=validation_dataset,
                    callbacks=[callback])
```

```python
plt.figure(figsize=(8,8))
ax=plt.subplot(2,1,1)
plt.plot(history.history['val_accuracy'],label='val_accuracy')
plt.plot(history.history['accuracy'],label='accuracy')
plt.ylabel("Accuracy(%)")
plt.grid()
plt.legend()

ax=plt.subplot(2,1,2)
plt.plot(history.history['val_loss'],label='val_loss')
plt.plot(history.history['loss'],label='loss')
plt.ylabel("LOSS")
plt.xlabel("Epoch")
plt.grid()
plt.legend()
plt.show()
```

```python
def unfreeze(base_model):
    base_model.trainable = True
    # Let's take a look to see how many layers are in the base model
    print("Number of layers in the base model: ", len(base_model.layers))

    # Fine-tune from this layer onwards
    fine_tune_at = 100

    # 冻结了前面的100层，从101开始到末尾可以训练，这样提高了拟合的层数，应该可以获得更好的精度
    # Freeze all the layers before the `fine_tune_at` layer
    for layer in base_model.layers[:fine_tune_at]:
        layer.trainable = False

    optimizer = tf.keras.optimizers.Adam(lr=base_learning_rate / 10)
    model.compile(optimizer=optimizer,
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model


mobilenet_unfreeze_model = unfreeze(base_model)
```

```python
fine_tuning_hist = mobilenet_unfreeze_model.fit(
    train_dataset,
    epochs=epochs_num_fine_tuning,
    validation_data=validation_dataset)
```

```python
plt.figure(figsize=(8,8))
ax=plt.subplot(2,1,1)
plt.plot(fine_tuning_hist.history['val_accuracy'],label='val_accuracy')
plt.plot(fine_tuning_hist.history['accuracy'],label='accuracy')
plt.ylabel("Accuracy(%)")
plt.grid()
plt.legend()

ax=plt.subplot(2,1,2)
plt.plot(fine_tuning_hist.history['val_loss'],label='val_loss')
plt.plot(fine_tuning_hist.history['loss'],label='loss')
plt.ylabel("LOSS")
plt.xlabel("Epoch")
plt.grid()
plt.legend()
plt.show()
```
