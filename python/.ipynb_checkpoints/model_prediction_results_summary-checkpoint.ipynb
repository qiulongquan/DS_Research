### EfficientNet_B0_model预测结果

```python
image.shape
```

```python
plt.figure(figsize=(15, 15))
# 设定空白处 间隔大小
plt.subplots_adjust(wspace=0.4, hspace=0.6)
count = 0
for i in range(BATCH_SIZE):
    ax = plt.subplot(8, 4, i + 1)
    plt.imshow(image[i].numpy().astype("uint8"))
    pred = efficientnet_model.predict(np.expand_dims(image[i], 0))
    Correct_value = label[i].numpy().astype('uint8').argmax()
    Predict_value = pred[0].argmax()
    plt.title("Correct value:{}\n Predict value:{}".format(
        Correct_value, Predict_value))
    if Correct_value == Predict_value:
        count += 1
    plt.axis("off")
correct_rate['efficientnet_model'] = round(count * 1.0 / BATCH_SIZE * 100, 2)
print("Efficientnet_model Correct rate={}%".format(
    correct_rate['efficientnet_model']))
```

```python
plt.figure(figsize=(15, 15))
# 设定空白处 间隔大小
plt.subplots_adjust(wspace=0.4, hspace=0.6)
count = 0
for i in range(BATCH_SIZE):
    ax = plt.subplot(8, 4, i + 1)
    plt.imshow(image[i].numpy().astype("uint8"))
    pred = mobilenetv2_model.predict(np.expand_dims(image[i], 0))
    Correct_value = label[i].numpy().astype('uint8').argmax()
    Predict_value = pred[0].argmax()
    plt.title("Correct value:{}\n Predict value:{}".format(
        Correct_value, Predict_value))
    if Correct_value == Predict_value:
        count += 1
    plt.axis("off")
correct_rate['mobilenetv2_model'] = round(count * 1.0 / BATCH_SIZE * 100, 2)
print("Mobilenetv2_model Correct rate={}%".format(
    correct_rate['mobilenetv2_model']))
```

### Baseline_model预测结果

```python
real_label=baseline_label.numpy()
```

```python
predictions = baseline_model.predict_on_batch(baseline_image).flatten()
```

```python
pred_result=[]
for i in range(BATCH_SIZE):
    pred_result.append(predictions[i*class_num:(i+1)*class_num].argmax())
```

```python
print(classification_report(real_label,pred_result,target_names=classes))
```

```python
cm = confusion_matrix(real_label,pred_result)
# print(cm)
sns.heatmap(cm, annot=True, cmap='Blues')
plt.title("予測結果")
plt.ylabel("実際結果")
plt.show()
```

```python
plt.figure(figsize=(15,15))
# 设定空白处 间隔大小
plt.subplots_adjust(wspace=0.4, hspace=0)
count=0
for i in range(BATCH_SIZE):
    ax=plt.subplot(4,8,i+1)
    plt.imshow(baseline_image[i].numpy().astype('uint8'))
    plt.axis('off')
    Correct_value=baseline_label[i].numpy().astype('uint8')
    Predict_value=pred_result[i]
    plt.title("Correct value:{}\n Predict value:{}".format(Correct_value,Predict_value))
    if Correct_value == Predict_value:
        count += 1
correct_rate['baseline_model'] = round(count * 1.0 / BATCH_SIZE * 100, 2)
print("Baseline_model Correct rate={}%".format(correct_rate['baseline_model']))
```
