### Base EfficientNet model perform transfer learning

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import tensorflow as tf
import keras
import pickle
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image_dataset_from_directory
from keras import models
from keras.models import load_model
from sklearn.metrics import accuracy_score
from keras import optimizers
from keras.preprocessing.image import ImageDataGenerator
```

```python
if os.getcwd().split('/')[-1] == "efficientnet_model":
    os.chdir("..")
PATH = os.path.join(os.getcwd(), 'hand_sign_digit_data')

train_dir = os.path.join(PATH, 'train')
validation_dir = os.path.join(PATH, 'validation')

classes = [
    'zero', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight',
    'nine'
]

class_num = len(classes)
BATCH_SIZE = 32
epochs_num = 50
epochs_num_fine_tuning = 15
IMG_SIZE = (224, 224)
```

```python
train_dataset = image_dataset_from_directory(train_dir,
                                             shuffle=True,
                                             batch_size=BATCH_SIZE,
                                             image_size=IMG_SIZE)
```

```python
validation_dataset = image_dataset_from_directory(validation_dir,
                                                  shuffle=True,
                                                  batch_size=BATCH_SIZE,
                                                  image_size=IMG_SIZE)
```

```python
val_batches = tf.data.experimental.cardinality(validation_dataset)
test_dataset = validation_dataset.take(val_batches//5)
validation_dataset = validation_dataset.skip(val_batches//5)

print('Number of validation batches: %d' % tf.data.experimental.cardinality(validation_dataset))
print('Number of test batches: %d' % tf.data.experimental.cardinality(test_dataset))
```

### Data Augmentation processing

```python
# 数据扩张需要注意，不要设定的过多，这样会降低准确率，拟合会很费劲，下面的4个设定使用就可以了
# 水平翻转，旋转，图片上下shift10%，左右shift10%，独立的调整每个图片每个通道的对比度
data_augmentation=tf.keras.Sequential([
#     水平 垂直 翻转
    tf.keras.layers.experimental.preprocessing.RandomFlip(mode='horizontal'),
#     旋转
    tf.keras.layers.experimental.preprocessing.RandomRotation(0.1),
#     缩放
#     tf.keras.layers.experimental.preprocessing.RandomZoom(-0.1, 0.1),
    #     图片上下shift10%，左右shift10%
    tf.keras.layers.experimental.preprocessing.RandomTranslation(height_factor=0.1, width_factor=0.1),
#     独立的调整每个图片每个通道的对比度
    tf.keras.layers.experimental.preprocessing.RandomContrast(factor=0.1),
])
```

```python
image, label = next(iter(validation_dataset))
len(image),label
```

```python
plt.figure(figsize=(15,15))
# 设定空白处 间隔大小
plt.subplots_adjust(wspace=0.2, hspace=0)
for i in range(BATCH_SIZE):
    ax=plt.subplot(4,8,i+1)
    im=image[i].numpy().astype('uint8')
    im=np.expand_dims(im,0)
    data_aug=data_augmentation(im)
    plt.imshow(data_aug[0])
    plt.title(label[i].numpy().astype('uint8'))
    plt.axis('off')
```

### One-hot / categorical encoding
把label标签的1维度值变成分类的10维度值

```python
def input_preprocess(image, label):
    label = tf.one_hot(label, class_num)
    return image, label


train_dataset = train_dataset.map(
    input_preprocess, num_parallel_calls=tf.data.experimental.AUTOTUNE)

train_dataset = train_dataset.prefetch(tf.data.experimental.AUTOTUNE)
validation_dataset = validation_dataset.map(input_preprocess)
```

### Create model with pre-trained convnets

```python
from tensorflow.keras.layers.experimental import preprocessing
from tensorflow.keras.applications import EfficientNetB0
import tensorflow


def build_model(num_classes):
    inputs = tensorflow.keras.layers.Input(shape=(IMG_SIZE[0], IMG_SIZE[1], 3))
    x = data_augmentation(inputs)
    model = EfficientNetB0(include_top=False,
                           input_tensor=x,
                           weights="imagenet",
                           classes=class_num,
                           classifier_activation="softmax",
                           drop_connect_rate=0.4)

    # Freeze the pretrained weights
    model.trainable = False

    # Rebuild top
    x = tensorflow.keras.layers.GlobalAveragePooling2D(name="avg_pool")(
        model.output)
    x = tensorflow.keras.layers.BatchNormalization()(x)

    top_dropout_rate = 0.2
    x = tensorflow.keras.layers.Dropout(top_dropout_rate,
                                        name="top_dropout")(x)
    outputs = tensorflow.keras.layers.Dense(num_classes,
                                            activation="softmax",
                                            name="pred")(x)

    # Compile
    model = tf.keras.Model(inputs, outputs, name="EfficientNet")
    optimizer = tf.keras.optimizers.Adam(learning_rate=1e-2)
    model.compile(optimizer=optimizer,
                  loss="categorical_crossentropy",
                  metrics=["accuracy"])
    return model
```

```python
strategy = tf.distribute.MirroredStrategy()
with strategy.scope():
    efficientnet_model = build_model(num_classes=class_num)
```

```python
callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=3)

hist=efficientnet_model.fit(
    train_dataset,
    epochs=epochs_num,
    validation_data=validation_dataset,
#     callbacks=[callback]
)
```

```python
plt.figure(figsize=(8,8))
ax=plt.subplot(2,1,1)
plt.plot(hist.history['val_accuracy'],label='val_accuracy')
plt.plot(hist.history['accuracy'],label='accuracy')
plt.ylabel("Accuracy(%)")
plt.grid()
plt.legend()

ax=plt.subplot(2,1,2)
plt.plot(hist.history['val_loss'],label='val_loss')
plt.plot(hist.history['loss'],label='loss')
plt.ylabel("LOSS")
plt.xlabel("Epoch")
plt.grid()
plt.legend()
plt.show()
```

### 解冻一部分层，然后进行微调训练 fine tuning

```python
def unfreeze(model):
    for layer in model.layers[-20:]:
#         BathcNormalization层不能解冻，再次训练会导致精度降低
        if not isinstance(layer, tf.keras.layers.BatchNormalization):
            print(layer.name)
            layer.trainable = True
    optimizer = tf.keras.optimizers.Adam(lr=1e-3)
    model.compile(optimizer=optimizer,
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    return model


efficientnet_unfreeze_model = unfreeze(efficientnet_model)
```

```python
fine_tuning_hist=efficientnet_unfreeze_model.fit(
    train_dataset,
    epochs=epochs_num_fine_tuning,
    validation_data=validation_dataset
)
```

```python
plt.figure(figsize=(8,8))
ax=plt.subplot(2,1,1)
plt.plot(fine_tuning_hist.history['val_accuracy'],label='val_accuracy')
plt.plot(fine_tuning_hist.history['accuracy'],label='accuracy')
plt.ylabel("Accuracy(%)")
plt.grid()
plt.legend()

ax=plt.subplot(2,1,2)
plt.plot(fine_tuning_hist.history['val_loss'],label='val_loss')
plt.plot(fine_tuning_hist.history['loss'],label='loss')
plt.ylabel("LOSS")
plt.xlabel("Epoch")
plt.grid()
plt.legend()
plt.show()
```
