### 原始模型，没有使用迁移模型

```python
# 查看系统path里面是否包含了python38的目录，如果没有执行下面的命令把python38的地址加入到path里面
import os
os.sys.path
```

```python
import tensorflow
import cv2
import datetime
from tensorflow import keras  # 这句话很重要，通过它可以导入keras
from keras import optimizers
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten

from tensorflow.keras.layers import Convolution2D, MaxPooling2D
from tensorflow.keras.callbacks import EarlyStopping
import sys
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
import matplotlib.image as img
import requests
from io import BytesIO
from PIL import Image
import numpy as np

%matplotlib inline

IMG_SIZE = 32
model_path = '/workspace/model'
pic_path = '/workspace/picture'
files_name_list = []

(X_train, y_train), (X_test,
                     y_test) = tensorflow.keras.datasets.cifar10.load_data()
```

```python
print(sys.executable)
```

```python
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
X_train.shape[1:],type(X_train)
```

```python
X_train.dtype
```

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
print("X_train dtype=", X_train.dtype)
X_train /= 255.0
X_test /= 255.0
```

```python
y_train = tensorflow.keras.utils.to_categorical(y_train, 10)
y_test = tensorflow.keras.utils.to_categorical(y_test, 10)
```

```python
X_train.shape[1:]
```

```python
model = Sequential()

model.add(
    Convolution2D(32, (3, 3),
                  padding='same',
                  activation='relu',
                  input_shape=X_train.shape[1:]))
model.add(Convolution2D(
    32,(3, 3),
    activation='relu',
))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Convolution2D(64, (3, 3), activation='relu', padding='same'))
model.add(Convolution2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))

# コンパイル
model.compile(loss='categorical_crossentropy',
              optimizer=optimizers.Adam(),
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(X_train,
                    y_train,
                    batch_size=100,
                    epochs=50,
                    verbose=1,
                    validation_split=0.1,
                    callbacks=[EarlyStopping()])
```

```python
score = model.evaluate(X_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
def prepare(file):
    img_array = cv2.imread(file)
    new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
    predictdata = np.expand_dims(new_array, 0)
    #     print(predictdata.dtype)
    #     predictdata /= 255.0
    return predictdata
```

```python
#os.walk方法获取当前路径下的root（所有路径）、dirs（所有子文件夹）、files（所有文件）
files_name_list = []


def get_files_name(pic_path):
    for root, dirs, files in os.walk(pic_path):
        for file in files:
            #print file.decode('gbk')    #文件名中有中文字符时转码
            if os.path.splitext(file)[1] == '.jpg':
                #                 print(os.path.join(root, file))
                files_name_list.append(os.path.join(root,
                                                    file))  #将所有的文件名添加到L列表中
    return files_name_list  # 返回L列表
```

```python
# 狗对应的是5，猫是3，鸟是2，卡车是9，飞机是0
files_name_list = get_files_name(pic_path)
for file_path in files_name_list:
    print(file_path)
    prediction = load_model.predict([prepare(file_path)])
    print(np.argmax(prediction))
```
